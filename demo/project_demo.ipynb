```python
from pathtracer import *
import time

# Configure the pathtracer quality settings
p = Pathtracer(36, 48)
p.rays_per_pixel = 1
p.depth = 3

# Load a scene with a single white plane
p.load_from_file("scenes/light_plane.json")
```

```python
# Run purely in software

t0 = time.time()
p.render_scene_in_software()
t1 = time.time()

print()
print(f"Took {t1 - t0:.1f} seconds.")

plt.imshow(p.usable_pixel_array())
plt.show()
```

```python
# Run with hardware acceleration

p.init_hardware()
print("Overlay done.")

t0 = time.time()
p.render_scene_in_hardware()
t1 = time.time()

print()
print(f"Took {t1 - t0:.1f} seconds.")

plt.imshow(p.usable_pixel_array())
plt.show()
```

```python
# Load a more complex scene
p.load_from_file("scenes/cornell_box_tri.json")
```

```python
# Run with hardware acceleration

t0 = time.time()
p.render_scene_in_hardware()
t1 = time.time()

print()
print(f"Took {t1 - t0:.1f} seconds.")

plt.imshow(p.usable_pixel_array())
plt.show()
```

```python
# Reconfigure the pathtracer quality settings for higher quality
p = Pathtracer(360, 480)
p.rays_per_pixel = 2
p.depth = 3
p.load_from_file("scenes/cornell_box_tri.json")

# Run again with hardware acceleration

p.init_hardware()

t0 = time.time()
p.render_scene_in_hardware()
t1 = time.time()

print()
print(f"Took {t1 - t0:.1f} seconds.")

plt.imshow(p.usable_pixel_array())
plt.show()

p.save_rendered_scene('hardware_result.png')
```
